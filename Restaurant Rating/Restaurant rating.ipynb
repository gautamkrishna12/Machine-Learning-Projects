```python
plt.figure(figsize=(4,3))
correlation = df[['Latitude', 'Longitude','Country Code','Aggregate rating']].corr()
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```

```python
corr_lat_rating = correlation.loc['Latitude', 'Aggregate rating']
print(f"* Correlation between Latitude and Aggregate rating: {corr_lat_rating}")

corr_long_rating = correlation.loc['Longitude', 'Aggregate rating']
print(f"\n* Correlation between Longitude and Aggregate rating: {corr_long_rating}")

corr_country_code= correlation.loc['Country Code','Aggregate rating']
print(f"\n* Correlation between Country code and Aggregate rating: {corr_country_code}")
```


Level 2
Task 1: Table Booking and Online Delivery


```python
#percentage of restaurants that offer table booking and online delivery

resturants=len(df[(df['Has Table booking']=='Yes') & (df['Has Online delivery']=='Yes')])*100/len(df)
print(f"* Percentage of restaurants that offer table booking and online delivery : {resturants:.2f} %")
```

```python
#Ratings of restaurants with table booking and those without
plt.figure(figsize=(6,4))
plt.subplot(121,)
df[df['Has Table booking']=='Yes']['Aggregate rating'].plot(kind='hist',bins=30)
plt.title('With Table Booking')
plt.subplot(122)
plt.title('without Table Booking')
df[df['Has Table booking']=='No']['Aggregate rating'].plot(kind='hist',bins=30)
plt.tight_layout()
```

```python
avg1=np.mean(df[df['Has Table booking']=='No']['Aggregate rating'])
print(f'* Average rating for resturants without table booking : {avg1:.2f}')

avg2=np.mean(df[df['Has Table booking']=='Yes']['Aggregate rating'])
print(f'\n* Average rating for resturants with table booking : {avg2:.2f}')
```


Task 2: Price Range Analysis


```python
#Most common price range
df['Price range'].value_counts().head(1)
```

```python
#Average rating for each price range.
print("Average rating for each price range.\n")
Rating_on_price=df[['Aggregate rating','Price range']].groupby("Price range").mean()
Rating_on_price
```

```python
max_avg_rating=round(Rating_on_price['Aggregate rating'].max(),1)
color=df[df['Aggregate rating']==max_avg_rating]['Rating color'].iloc[0]
print(f"THe maximum average rating : {max_avg_rating}\n\nRating color : {color}.")
```
